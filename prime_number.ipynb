```python
def prime_number(num):
    if num == 2 or num == 3 or num == 5 :
        return("Prime Number")    
    elif num%2 == 0:
        return("Not Prime Number")
    elif num%3 == 0:
        return("Not Prime Number")
    elif num%5 == 0:
        return("Not Prime Number")
    else:
        return("Prime Number")
```

```python
for i in range(120):
    print(i,prime_number(i))
```
